# Activation Functions
 - Linear activation function <br/>
 - binary step <br/>
 - logistic or sigmoid <br/>
 - tanh <br/>
 - arctan <br/>
 - Rectified Linear Unit (ReLu) <br/>
 - Leaky Rectified Unit ReLu <br/>

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(-10, 10, 0.1)
```

```python
def binary_step(x):
    res = []
    for i in x:
        if i > 0:
            res.append(1)
        else:
            res.append(0)
    return res
            
def sigmoid(x):
    return 1 / 1 + np.exp(x)

def tanh(x):
    return 2 / ((1 + (np.exp(x)**2)) - 1)

def arctan(x):
    res = []
    for i in x:
        res.append(math.atan(i))
    return res

def ReLu(x):
    res = []
    for i in x:
        if i <= 0:
            res.append(0)
        else:
            res.append(i)
    return res
 
def leakyReLu(x):
    res = []
    for i in x:
        if i <= 0:
            res.append(0.1 * i)
        else:
            res.append(i)
    return res
```

```python
y1 = x
y2 = binary_step(x)
y3 = sigmoid(x)
y4 = tanh(x)
y5 = arctan(x)
y6 = ReLu(x)
y7 = leakyReLu(x)
```

```python
plt.figure(figsize = (30, 40))
plt.subplot(4, 2, 1)
plt.plot(x, y1)
plt.subplot(4, 2, 2)
plt.plot(x, y2)
plt.subplot(4, 2, 3)
plt.plot(x, y3)
plt.subplot(4, 2, 4)
plt.plot(x, y4)
plt.subplot(4, 2, 5)
plt.plot(x, y5)
plt.subplot(4, 2, 6)
plt.plot(x, y6)
plt.subplot(4, 2, 7)
plt.plot(x, y7)
```
