# Plotting Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.arange(0, 10)
y1 = x1
plt.plot(x1, y1, label = 'y = x', c = 'r', lw = 4)
plt.xlabel('x->')
plt.ylabel('y->')
plt.title('Graph of y = x')
plt.legend()
```

```python
x2 = np.arange(0, 10)
y2 = 2 * x2 + 5
plt.plot(x2, y2, label = 'y = 2x + 5', marker = '*', markersize = 20, markerfacecolor = 'r')
plt.legend()
```

```python
x3 = np.arange(0, 10)
y3 = x3 * x3
plt.plot(x3, y3, c = 'g', lw = 3, marker = 'd', markerfacecolor = 'b', markersize = 10)
```
