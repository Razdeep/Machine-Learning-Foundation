# Plotting Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y1 = x ** 2
y2 = x ** 3
y3 = x
y4 = x ** 0.5
plt.subplot(2, 2, 1)
plt.title('Subplots #1')
plt.plot(x, y1)
plt.subplot(2, 2, 2)
plt.plot(x, y2, c = 'g')
plt.subplot(2, 2, 3)
plt.plot(x, y3, c = 'r')
plt.subplot(2, 2, 4)
plt.plot(x, y4, c = 'b')
```
